```python
import pandas as pd
import matplotlib.pylab as plt
%matplotlib inline
```

```python
df=pd.read_csv('salaries.csv')
df
```

```python
x=df.drop('salary_more_then_100k',axis='columns')
y=df.salary_more_then_100k
```

```python
from sklearn.preprocessing import LabelEncoder
com=LabelEncoder()
```

```python
x['com_n']=com.fit_transform(x['job'])
x['job_n']=com.fit_transform(x['job'])
x['deg_n']=com.fit_transform(x['degree'])
```

```python
x
```

```python
x_df=x.drop(['company','job','degree'],axis='columns')
```

```python
x_df
```

```python
from sklearn import tree
from sklearn.tree import plot_tree
```

```python
model=tree.DecisionTreeClassifier()
```

```python
model.fit(x_df,y)
```

```python
model.score(x_df,y)
```

```python
model.predict([[1,0,1]])
```

```python
plt.figure(figsize=(20,10))
plot_tree(model,filled=True)
```

```python
plt.figure(figsize=(20,10))
plot_tree(model,filled=True)
```

```python
df1=pd.read_csv('titanic.csv')
df1.head()
```

```python
df1.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
```

```python
df1.head()
```

```python
x=df1.drop('Survived',axis='columns')
y=df1.Survived
```

```python
x.replace({'male':0,'female':1},inplace=True)
```

```python
x.head()
```

```python
 x.isnull().sum()
```

```python
x['Age']=x['Age'].fillna(x.Age.mean())
```

```python
 x.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree
```

```python
m=DecisionTreeClassifier()
```

```python
m.fit(X_train,y_train)
```

```python
m.score(X_test,y_test)
```

```python
m.predict(X_test)
```

```python
plt.figure(figsize=(80,40))
plot_tree(m,filled=True)
```
